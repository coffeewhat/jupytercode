```python
im_dir_path = '/Users/zhangxiao/Desktop/测测试数据/交通数据'
im_dir_names = os.listdir(im_dir_path)
```

```python
im_dir_names.sort()
im_dir_names.remove('标注案例.xlsx')
```

```python
print(im_dir_names)
```

```python
im_paths=[]
for im_dir_name in im_dir_names:
    dir = os.path.join(im_dir_path, im_dir_name) # 其中一个文件夹路径
    im_path = list(paths.list_images(dir))
    im_path.sort()
    im_paths += im_path
```

```python
print(im_paths)
```

```python
len(im_paths)
```

```python
i = 0
for im_path in im_paths:
    pathSplit = os.path.split(im_path)
    os.rename(im_path, os.path.join(pathSplit[0], l[i] + '_' + str(i) + '_' + pathSplit[1]))
    i += 1
```
