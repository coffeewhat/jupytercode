```python
def aug_apply(img, aug, times):
    return [aug(img) for _ in range(times)]    
```

```python
sample_num = len(Labels)
times = 8
sample_C = 3
sample_H = 480
sample_W = 640
datasets = torch.zeros((sample_num*times, sample_C, sample_H, sample_W))
targets = torch.zeros((sample_num*times))
aug = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor()])
i = 0
j = 0
for im_path in im_paths:
    img = Image.open(im_path)   
    Y = aug_apply(img, aug, times)
    for y in Y:
        datasets[i] = y
        targets[i] = Labels[j]
        i+=1
    j+=1    
```

```python
print(i, j)
```

```python
datasets[0].shape # C H W
```

```python
Datasets = torch.utils.data.TensorDataset(datasets, targets)
```

```python
batch_size = 256
test_split = 0.2
shuffle_dataset = True
random_seed = 1
datasets_size = len(Datasets)
indices = list(range(datasets_size))
split = int(np.floor(test_split * datasets_size))
if shuffle_dataset :
    np.random.seed(random_seed)
    np.random.shuffle(indices)
train_indices, test_indices = indices[split:], indices[:split]
```

```python
train_sampler = torch.utils.data.SubsetRandomSampler(train_indices)
test_sampler = torch.utils.data.SubsetRandomSampler(test_indices)

train_iter = torch.utils.data.DataLoader(Datasets, batch_size=batch_size, 
                                           sampler=train_sampler)
test_iter = torch.utils.data.DataLoader(Datasets, batch_size=batch_size,
                                                sampler=test_sampler)
```

```python
for X, y in train_iter:
    print(X, y)
    break
```

```python
joblib.dump((train_iter, test_iter), 'data.pkl')
```
