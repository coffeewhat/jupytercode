```python
plt.figure()
plt.text( 0.5, 0.5, 'Figure', ha='center', 
          va='center', size=20, alpha=.5 )
plt.xticks([]), plt.yticks([])
plt.show()
```

用 plt.text() 函数，其参数解释如下：

第一、二个参数是指横轴和纵轴坐标

第三个参数字符是指要显示的内容

ha, va 是横向和纵向位置

size 设置字体大小

alpha 设置字体透明度 (0.5 是半透明)

```python
plt.figure()
plt.plot( [0,1],[0,1] ) #前两个参数分别是 x 和 y
plt.show()
```

```python
plt.subplot(2, 1, 1)# 一幅图中的第一子图
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot(2, 1, 1)', ha='center', va='center', size=20, alpha=0.5)

plt.subplot(2, 1, 2)
plt.xticks([]), plt.yticks([])# 一副图中的第二个子图
plt.text(0.5, 0.5, 'subplot(2, 1, 2)', ha='center', va='center', size=20, alpha=0.5)

plt.show()
```

# 因此坐标系和子图是不是同样的概念？

在绝大多数情况下是的，两者有一点细微差别：

子图在母图中的网格结构一定是规则的

坐标系在母图中的网格结构可以是不规则的

```python
plt.subplot(1, 2, 1)# 一幅图中的第一子图
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot(2, 1, 1)', ha='center', va='center', size=15, alpha=0.5)

plt.subplot(1, 2, 2)
plt.xticks([]), plt.yticks([])# 一副图中的第二个子图
plt.text(0.5, 0.5, 'subplot(2, 1, 2)', ha='center', va='center', size=15, alpha=0.5)

plt.show()
```

## 两种生成单个坐标系的方法

1.同时生成图和坐标系

```python
fig, ax = plt.subplots()
ax.set(xticks=[], yticks=[])
s = "Style 1\n\nfig, ax = plt.subplots()\nax.plot()"
ax.text(0.5, 0.5, s, ha='center', va='center', size=15, alpha=0.5)
```

2.先生成图，再生成坐标系

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set(xticks=[], yticks=[])
s = 'Style 2\n\nfig.subplots()\nax=fig.add_subplot(1, 1, 1)\nax.plot()'
ax.text(0.5, 0.5, s, ha='center', va='center', size=16, alpha=0.5)
```
